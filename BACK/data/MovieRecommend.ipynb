```python
import pandas as pd
import numpy as np
import warnings; warnings.filterwarnings('ignore')
```

```python
movies =pd.read_json('../movies/fixtures/movies_data_1000.json')

print(movies.shape)
movies.head(1)
```

```python
movies.info()
```

```python
genre_list = {28: '액션', 12: '모험', 16: '애니메이션', 35: '코미디', 80: '범죄', 99: '다큐멘터리', 18: '드라마', 10751: '가족', 14: '판타지', 36: '역사', 27: '공포', 10402: '음악', 9648: '미스터리', 10749: '로맨스', 878: 'SF', 10770: 'TV 영화', 53: '스릴러', 10752: '전쟁', 37: '서부'}
```

```python
movies['genres_literal'] = np.NaN

for i in range(1000):
    pk_list = movies['genres'][i]
    movies['genres_literal'][i] = ' '.join([genre_list.get(pk) for pk in pk_list])
```

```python
movies.head()
```

```python
movies_df = movies
```

```python
from sklearn.feature_extraction.text import CountVectorizer

count_vect = CountVectorizer(min_df=0, ngram_range=(1,2)) 
#ngram_range 단어 최대 2개까지 이용해서 피처를 생성, 단어1개 피처 단어 2개를 조합한 피처가 생성
count_vect.fit(movies_df['genres_literal'])
```

```python
count_vect.get_feature_names()
```

```python
count_vect.vocabulary_ # 각 feature의 인덱스
```

```python
# 장르변환
genre_mat = count_vect.transform(movies_df['genres_literal'])
print(genre_mat.shape)
```

```python
type(genre_mat)
```

```python
genre_mat.toarray()
```

```python
pd.DataFrame(genre_mat.toarray(), columns=count_vect.get_feature_names())
```

```python
# 사이킷런의 cosine_similarity() 함수를 이용해 코사인 유사도 계산

from sklearn.metrics.pairwise import cosine_similarity

genre_sim = cosine_similarity(genre_mat, genre_mat)

print(genre_sim.shape) # 유사도 행렬의 크기
print(genre_sim[:10]) # 2개행만 확인
```

```python
def find_sim_movie(df, sorted_ind, title_name, top_n=10):
    
    # 인자로 전달된 title에 해당되는 데이터 추출
    title_movie = df[df['title'] == title_name]
    
    # title_named을 가진 DataFrame의 index 객체를 ndarray로 반환하고 
    # sorted_ind 인자로 입력된 genre_sim_sorted_ind 객체에서 
    # 유사도 순으로 top_n 개의 index 추출
    title_index = title_movie.index.values
    similar_indexes = sorted_ind[title_index, :(top_n)]
    
    # 추출된 top_n index들 출력. top_n index는 2차원 데이터 임. 
    #dataframe에서 index로 사용하기 위해서 1차원 array로 변경
    print(similar_indexes)
    similar_indexes = similar_indexes.reshape(-1)
    
    # index에 해당되는 데이터프레임 반환
    return df.iloc[similar_indexes]
```

```python
similar_movies = find_sim_movie(movies_df, genre_sim_sorted_ind, '블랙 아담',10)
```
